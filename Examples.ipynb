```python
# Computes the Lagrangians in the pillowcase associated to rational tangles
T = RatTangle(3)
S = RatTangle(5)
R = RatTangle(2)

#Note: RatTangle(p) actually calculates the character variety of T_{1/p} which follows the pretzel knot convention. This will be changed soon.
```

```python
S.show()
```

```python
# Compute the  Lagrangians associated to T+S (currently only works if T and S are rational tangles, soon to be expanded to arbitrary tangles)
C = CRat([T,S])
C.show()
```

```python
# For tangle sums, the result can be perturbed (up to homotopy) by a certain perturbation curve defined in [Smith]
C.perturb(1/100)
C.show()
```

```python
# Character varieties can be modified in the way described in [CHKK], which is associated to adding an 'earring' to the tangle
# Arc components are replaced by figure-8s and the parameter dictates the size of the figure-8 in a way
#Resturns a 'distinguished path' which is useful in future steps
dp = R.earring(1/100)
```

```python
# A knot can be analyzed by splitting it into two tanlges and analyzing the pair of associated Lagrangians in the pillowcase.
# For example,the pretzel knots P(2,3,5) can be decomposed into T_{1/2} and T_{1/3}+T_{1/5} by a Conway sphere
# The distpath parameter helps define a distinguished intersection

I = Intersection(C,R,distpath=dp)
I.show()
# I.ips is a list of intersection point
```

```python
# Intersections can be (relatively) graded according to [CHHK, Section 5]. Picking a smart decomposition of tangles where each character variety is connected allows one to get an absolute grading
# Color/size of intersection defines Z/4-grading (minus the signature of the knot)

I.grade()
I.showgr()
# I.grips is a length 4 list of intersection points by grading
```

```python
I.grgens() # gives the number of intersections at each grading

# These generate CF(L_1, L_2). This is isomorphic to CI^\natural(K) and therefore bounds the singular instanton homology.

# chaincx(3,5,2) is a shortcut for running everything up until now and outputing I.grgens()
```

```python
# To compute the Lagrangian Floer homology, differentials are given by certain immersed bigons which don't contain the corner points.

I.findbigons() # Computes these differntials based on the algorithm from [Blank]

#Show all of the bigons and their lifts into R^2
for b in I.bigons:
    I.showbigon(b)
```

```python
print(I.homology()) # Computes the homology

I.showchains() # Shows the chain complex. The points in the north represent grading 0 generators. East are grading 1 etc. Then differnteials are given by the lines, with direction following the gradings. Color represents if its binary dihedral.
```
